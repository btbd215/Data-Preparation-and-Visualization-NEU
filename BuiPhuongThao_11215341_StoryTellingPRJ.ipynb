* Creating a graph for multiple currencies – perhaps you can compare their evolution

```python
chosen = ex_rate[['Time','[Danish krone ]','[UK pound sterling ]','[Iceland krona ]','[Thai baht ]','[Swedish krona ]','[Philippine peso ]','[Hong Kong dollar ]', '[Chinese yuan renminbi ]','[Japanese yen ]', '[Norwegian krone ]']]
```

```python
chosen = chosen.copy()
chosen['Time'] = pd.to_datetime(chosen['Time'])
```

```python
chosen['Time'] = chosen['Time'].dt.strftime('%Y')
```

```python
def fill(df):
    for i in df.columns:
        if i == 'Time':
            continue
        else:
            df = df.copy()
            df = df[df[i] != '-']
            df[i] = df[i].astype(float)
            mean = df[i].mean()
            df[i].fillna(value = mean, inplace = True)
    return df
```

```python
chosen = fill(chosen)
```

```python
cop = ex_rate.copy()
cop = fill(cop)
```

```python
cop1 = cop[cop.columns[1:]]
#cop1
```

```python
cm = cop1.max()
```

```python
nho = []
for i in cm.index:
    if cm[i] > 4 and cm[i] < 10:
        nho.append(i) 
```

```python
nho
```

```python
nhodf = cop[nho]
nhodf = nhodf.copy()
nhodf['Time'] = cop['Time']
nhodf['Time'] = pd.to_datetime(nhodf['Time']).dt.strftime('%Y')
#chosen['Time'] = chosen['Time'].dt.strftime('%Y')
#nhodf
```

```python
fig,ax5 = plt.subplots(figsize = (25,10))
ax5.plot(nhodf.groupby('Time')[nhodf.columns[:8]].apply(lambda x: x.mean()),lw = 4, label = [i for i in nhodf.columns[:8]])
ax5.spines[['top','right']].set_visible(False)
ax5.legend()
ax5.text(-1,8,'Changes in currency whose rate is under 8 each year', weight = 'bold', size = 17)
```
