```python
#Create a new dataframe named frequency_df
frequency_df = pd.DataFrame({'CustomerID': data['CustomerID'].drop_duplicates()})
```

```python
temp = data.groupby('CustomerID')[['InvoiceNo']].nunique()
temp
```

```python
frequency_df.columns = ['CustomerID', 'Frequency']
frequency_df
```

```python
data['TotalRevenue'] = data['Quantity'] * data['UnitPrice']
```

```python
monetary_df = data.groupby('CustomerID')['TotalRevenue'].sum().reset_index()
```

```python
monetary_df.columns = ['CustomerID', 'Monetary']
```

```python
recency_df = (data['InvoiceDate'].max() - data.groupby('CustomerID')['InvoiceDate'].max()).dt.days.reset_index()
recency_df.columns = ['CustomerID','Recency']
```

```python
recency_df
```

```python
customer_df = pd.DataFrame()
customer_df['CustomerID'] = data['CustomerID'].drop_duplicates()
customer_df = pd.merge(pd.merge(frequency_df,monetary_df,on='CustomerID'),recency_df,on='CustomerID')
customer_df
```

```python
quantiles = customer_df[['Frequency','Monetary','Recency']].quantile(q = [0.25,0.5,0.75,1])
quantiles
```

```python
quantiles = quantiles.to_dict()
```

```python
quantiles
```

```python
customer_df['r_quantile'] = pd.qcut(customer_df['Recency'], 4, labels = ['Q1', 'Q2', 'Q3', 'Q4'])
customer_df
```

```python
customer_df['Frequency'].unique()
```

```python
customer_df['f_quantile'] = pd.qcut(customer_df['Frequency'],4,labels = ['Q1','Q2','Q3'], duplicates = 'drop')
customer_df
```

```python
customer_df['m_quantile'] = pd.qcut(customer_df['Monetary'], 4, labels = ['Q1','Q2','Q3','Q4'])
customer_df
```

```python
def condition(x):
    if x == 'Q1':
        return 1
    elif x == 'Q2':
        return 2
    elif x == 'Q3':
        return 3
    else:
        return 4
```

```python
def recency_con(x):
    if x == 'Q1':
        return 4
    elif x == 'Q2':
        return 3
    elif x == 'Q3':
        return 2
    else:
        return 1
    
```

```python
customer_df['Monetary'] = customer_df['m_quantile'].apply(condition)
customer_df['Frequency'] = customer_df['f_quantile'].apply(condition)
customer_df['Recency'] = customer_df['r_quantile'].apply(recency_con)
```

```python
customer_df
```

```python
customer_df['RMF_Score'] = customer_df['Frequency'].astype(int) + customer_df['Monetary'].astype(int) + customer_df['Recency'].astype(int)
customer_df
```

```python
customer_df.sort_values(by = 'RMF_Score', axis = 0, ascending = True)
```

```python
score_count = customer_df.groupby('RMF_Score').count()['CustomerID']
```

```python
customer_df['Segment'] = np.where(customer_df["RMF_Score"] <= threshold[0], "low-value", np.where(customer_df["RMF_Score"] <= threshold[1], "mid-value", "high-value"))
```
