#### 2. Đổi tên cột từ định dạng camelcase sang snakecase. Ví dụ: dateCrawled ->date_crawled. Gợi ý: dùng df.columns

```python
data.columns
```

```python
def snake(name):
    new = ''
    for i in name:
        if i.isupper():
            new += '_'+i.lower()
        else:
            new += i
    return new
```

```python
data.columns = [snake(col) for col in data.columns]
```

#### 7. Đối với các features dạng numeric, fillna bởi giá trị median của feature đó. Còn đối với các features dạng categorical thì fillna bởi giá trị quan sát được nhiều nhất.

```python
null_check = data.isnull().sum()
null_check
```

```python
for i in null_check.index:
    if null_check[i] != 0:
        if data[i].dtypes != object:
            data[i].fillna(value = data[i].median(), inplace = True)
        else:
            data[i].fillna(value = data[i].value_counts().index[0],inplace = True)
```
